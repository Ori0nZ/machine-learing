```python
from sklearn.datasets import load_wine
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.metrics import adjusted_rand_score
```

```python
wine = load_wine()
for key in wine:
    print(key)
```

```python
X = wine.data[:,2]
print(X.shape)
```

```python
print (X)
```

```python

fig, axes = plt.subplots(1,1,figsize=(8,4))
axes = plt.scatter(X[:], wine.data[:,3])
plt.xlabel( "Unit", fontsize=14)
plt.ylabel("Value", fontsize=14)
```

```python
kmeans = KMeans(n_clusters = 3).fit(X.reshape(-1,1))
pred_label = kmeans.predict(X.reshape(-1,1))
kmeans.cluster_centers_
```

```python
print(kmeans.labels_) 
```

```python
plt.scatter(X[:],wine.data[:,3], c=kmeans.labels_, cmap='rainbow')  
plt.xlabel( "Unit", fontsize=14)
plt.ylabel("Value", fontsize=14)
```

```python
adjusted_rand_score(X, kmeans.labels_)
```
